```python
# Warning! This scraper takes approximately 5 seconds to get the price for each sku.
# Please be minfull of the execution time when adding multiple skus.
from bs4 import BeautifulSoup
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import random
import time
import requests


class scraper_hidroterra():
    # Initializing driver so it can be used by the other funtions.
    def __init__(self):
        self.options = webdriver.ChromeOptions()
        self.options.add_argument("--headless")
        self.to_driver = "C:/Program Files/ChromeDriver/chromedriver.exe"
        self.driver = webdriver.Chrome(self.to_driver, options=self.options)
        self.driver.close()
    
    #Funtion that opens driver            
    def open_driver(self, url):
        self.driver = webdriver.Chrome(self.to_driver, options=self.options)
        self.driver.get(url)
    
    #Function that performs a search for the selected sku using the search box on the webpage
    def search_product(self,sku):
        search_box = self.driver.find_element_by_name("searchTerm")
        search_box.send_keys(sku)
        search_box.send_keys(Keys.RETURN)
    
    #Funtion that finds the price in the webpage html and returns it in an appropiate format
    def get_price(self):
        page_source = self.driver.page_source
        page =  BeautifulSoup(page_source, "lxml")
        price = page.select("span[class='price']")
        price_text = price[0].text.strip()
        if len(price_text) == 0:
            return("ND")
        else:
            price_format = price_text[1:-2] + "." + price_text[-2:]
            return price_format
    
    #Function that generation a random wait time and page movement to avoid looking like a scraper
    def random_keypress_generator():
        keys = [Keys.ARROW_DOWN, Keys.ARROW_LEFT, Keys.ARROW_UP]
        action = np.random.choice(keys, p=[0.4, 0.4, 0.1])
        return action
    
    #Function that implements the random movements generated before and adds an idle wait time
    def wait(self):
        ran_number = np.random.choice
        for ran in range(ran_number(2)):
            keys = [Keys.ARROW_DOWN, Keys.ARROW_LEFT, Keys.ARROW_UP]
            action = np.random.choice(keys, p=[0.4, 0.4, 0.2])
            ActionChains(self.driver).send_keys(action).perform()
            time.sleep(random.randint(2,3))


#url used with this scraping program.
url = "https://www.homedepot.com.mx/"
# Products that will be scraped
skus = [337859,337875,337867,337962,334593,648675,648640,337946,933701,
        333892,152840,600480,338438,338462,338446,333005,714896,
        714888,714861,714845,933890,338403,59900,801035,550072,549948,
        549905,405719,405700,500520,127337,127333,127243]
skus = [337859,337875]
prices = {}
#Instantiating the scraper class
my_scraper = scraper_hidroterra()
#Opening the scraper
my_scraper.open_driver(url)
#For cicle that collects the prices and adds the to the dictionary "prices"
for sku in skus:
    my_scraper.search_product(str(sku))
    prices[sku] = my_scraper.get_price()
    my_scraper.wait()

print(prices)
```
